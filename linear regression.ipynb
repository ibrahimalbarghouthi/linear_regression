```ruby
require 'nyaplot'

g = -> (x, t, p) { (t - (x * p)) ** 2 }

gprime = -> (x, t, p) { 2 * x * ( p * x - t) }

f = -> (x) {x * 12}

def line(slope, x, b)
   slope * x + b
end

x = [1, 2, 3, 4, 5, 6, 7, 8]

t = x.map(&f)
iterate = (0..20).step(0.2)
m = iterate.map {|p| x.zip(t).inject(0) {|s, (x, t)| s += g.call(x, t, p) } }
m2 = iterate.map {|p| x.zip(t).inject(0) {|s, (x, t)| s += gprime.call(x, t, p) } }

plt = Nyaplot::Plot.new

lin = plt.add(:line, iterate.to_a, m)
p = 20

20.times do
  gradient = x.zip(t).inject(0) {|s, (x, t)| s += gprime.call(x, t, p) }
  point =  x.zip(t).inject(0) {|s, (x, t)| s += g.call(x, t, p) }
  
  lin = plt.add(:scatter, [p], [point]).color('rgb(215,48,39)')
  ln = line(gradient, -1, point)
  ln2 = line(gradient, 1, point)

  plt.add(:line, [p - 1, p + 1], [ln, ln2]).color('rgb(214,96,77)')
  p -= 0.001 * gradient

end
lin.color('rgb(215,48,39)')
plt
```
